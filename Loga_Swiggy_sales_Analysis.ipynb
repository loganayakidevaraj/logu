**Visulaization-Monthly Sales Trend – Shows how total sales fluctuate month by month.**

```python
import matplotlib.pyplot as plt
# Convert Order Date into datetime format
#df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract Month and Year
df['Month'] = df['Order Date'].dt.strftime('%Y-%m')

# Calculate Monthly Sales
monthly_sales = df.groupby('Month')['Price (INR)'].sum()
print(monthly_sales)

# Visualization
monthly_sales.plot(
    kind='line',
    marker='o'
   )
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.ylabel("Total Sales (INR)")
plt.show()
```

**Daily Sales Trend – Highlights order and revenue variations across days of the week.**

```python
df['Day'] = df['Order Date'].dt.day_name()

# Calculate Daily Sales
daily_sales = df.groupby('Day')['Price (INR)'].sum()

# Arrange weekdays properly
days_order = [
    'Monday',
    'Tuesday',
    'Wednesday',
    'Thursday',
    'Friday',
    'Saturday',
    'Sunday'
]

daily_sales = daily_sales.reindex(days_order)

# Display output
print(daily_sales)

# Visualization
daily_sales.plot(kind='bar')

plt.title("Daily Sales Trend")
plt.xlabel("Days of the Week")
plt.ylabel("Total Sales (INR)")

# Rotate x-axis labels
plt.xticks(rotation=45)

plt.show()
```

**Total Sales by Food Type (Veg vs Non-Veg) – Compares revenue contribution by cuisine type.**

```python
import numpy as np

df["Food Category"] = "Veg"

# Identify Non-Veg dishes
df.loc[
    df["Dish Name"].str.contains(
        "chicken|egg|fish|mutton|prawn|biryani",
        case=False,
        na=False
    ),
    "Food Category"
] = "Non-Veg"

# Total Sales
food_revenue = df.groupby(
    "Food Category"
)["Price (INR)"].sum()

# Display Output
print(food_revenue)

# Visualization
food_revenue.plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(8,6)
)

plt.title("Total Sales by Food Type")

plt.ylabel('')

plt.show()
```

```python
# Extract Quarter
df["Quarter"] = df["Order Date"].dt.quarter

# Quarterly Summary
quarter_summary = df.groupby("Quarter").agg({

    "Price (INR)": "sum",
    "Rating": "mean",
    "Rating Count": "sum"
})

# Rename columns
quarter_summary.columns = [
    "Total Sales",
    "Average Rating",
    "Total Orders"
]

# Display output
print(quarter_summary)

# Visualization
quarter_summary.plot(
    kind='bar',
    figsize=(10,6)
)

plt.title("Quarterly Performance Summary")
plt.xlabel("Quarter")
plt.ylabel("Values")

plt.xticks(rotation=0)


plt.show()
```

```python
top_cities = (
    df.groupby("City")["Price (INR)"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

# Display output
print(top_cities)

# Visualization
plt.figure(figsize=(10,5))

plt.bar(
    top_cities.index,
    top_cities.values
)

plt.title("Top 5 Cities by Sales")
plt.xlabel("City")
plt.ylabel("Total Sales (INR)")

plt.xticks(rotation=45)

plt.show()
```

```python
#Create Week column
df["Week"] = df["Order Date"].dt.isocalendar().week

# Weekly Sales
weekly_sales = df.groupby(
    "Week"
)["Price (INR)"].sum()

# Display output
print(weekly_sales)

# Visualization
weekly_sales.plot(
    kind='line',
    marker='o',
    figsize=(10,5)
)

plt.title("Weekly Trend Analysis")
plt.xlabel("Week Number")
plt.ylabel("Total Sales (INR)")

plt.grid(True)

plt.show()
```
